```python
import pandas as pd
import numpy as np

df = pd.read_csv('results.csv')
pd.set_option('display.max_columns', None)
df
```

```python
df = df.dropna()
df
```

```python
# Number of Tuples
print(len(df))
```

```python
# Number of Tournaments

df['tournament'].value_counts()
```

```python
import datetime as dt

#converts to timestamp
df['date'] = pd.to_datetime(df['date'])

df
```

```python
#Getting only the year from the dataframe and setting it into the variable year
year = pd.DatetimeIndex(df['date']).year
#intitalizing a counter for games played
count_game = 0
#looping through the year dataset and comparing them to see if they are in 2018. If so, add one to the counter
for i in year:
    if i == 2018:
        count_game +=1
count_game
```

```python
#Initialzing counters for the three results
home_wins = 0
home_losses = 0
home_draw = 0

#Checks to see if home score is greater than away score
total = df['home_score'] > df['away_score']
for i in total:
    if i == True:
        home_wins += 1
        
#Checks to see if home score is less than away score
total_losses = df['home_score'] < df['away_score']
for i in total_losses:
    if i == True:
        home_losses += 1
        
#Checks to see if home score is equal to away score
total_draws = df['home_score'] == df['away_score']
for i in total_draws:
    if i == True:
        home_draw += 1
        
print(f'Wins: {home_wins}\nLosses: {home_losses}\nDraws: {home_draw}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
y = np.array([home_wins, home_losses])
mylabels = ['Wins', 'Losses']
plt.pie(y, labels = mylabels)
plt.title('Home Wins vs Losses')
```
